## Цели и задачи

### Обоснование стратегии

Принципы, лежащие в основе технического анализа:

1. Начавшаяся тенденция, по всей вероятности, будет стремиться развиваться и далее, а не обращаться в собственную противоположность. 

2. Действующая тенденция будет развиваться до тех пор, пока не появятся фундаментальные факторы способные повлиять на направление движения цены.

Таким образом, наиболее оптимальное соотношение риск/доходность можно получить при работе по тренду (а не против него).
_____________

Основа тестируемой стратегии: 

1. определение начала тренда

2. открытие сделки в направлении тренда

3. определение окончания тренда и закрытие сделки.
______________________________

Одним из способов определить начало новой тенденции может быть индикатор RSI.

`RSI (Relative strength index, индекс относительной силы)` — индикатор технического анализа, показывающий соотношение положительных и отрицательных изменений цены финансового инструмента.

Индекс относительной силы может давать сигнал к покупке или продаже несколькими способами:

- пересечение 50 

- дивергенция 

- перекупленность и перепроданность 

- линии тренда.
 
Пересечение индикатором уровня 50 означает, что преобладающим становится усредненное движение либо вверх, либо вниз. Такое пересечение может свидетельствовать о начале тренда. 
______________________

Для следования за тенденцией часто используется `Скользящее среднее` : в данном случае было выбрано экспоненциально-сглаженное скользящее среднее(EMA), т к оно учитывает всю прошлую ценовую истрию, больше веса придает последним значениям, менее чувствительно к волатильности (относительно линейно-взвешенного ск.ср.).

Осциллятор MACD - производная от EMA, может помочь определить разворот тренда.

`Момент (Momentum)` – индикатор, отслеживающий рост или снижение скорости движения тренда. Он обычно достигает экстремальных значений раньше цен и может использоваться как сигнал для закрытия сделки.

### Создаем EMA с разными периодами

```python
# ф-ция для EMA, принимает перирод и сглаживание
# сглаживание smooth везде будет 2
def create_simple_ema(n, smooth):
    a = smooth/(1+n)
    ema = list()
    for i in list(range(0, df.shape[0])):
        if i == 0:
            ema.append(df['close'][i])
        else:
            ema_val = (df['close'][i] * a) + (ema[i-1] * (1-a))
            ema.append(ema_val)
    return ema
```

```python
# новый признак ema_{период}: 12, 24, 60, 120
df['ema_12'] = pd.Series(create_simple_ema(12, 2))
df['ema_24'] = pd.Series(create_simple_ema(24, 2))
df['ema_60'] = pd.Series(create_simple_ema(60, 2))
df['ema_120'] = pd.Series(create_simple_ema(120, 2))

# MACD между средней (период 60) и медленной (перирод 120) ЕМА 
df['MACD_60_120'] = df['ema_60'] - df['ema_120']
# MACD между средне-быстрой (период 24) и средней (перирод 60) ЕМА 
df['MACD_24_60'] = df['ema_24'] - df['ema_60']
# MACD между быстрой (период 12) и средней (перирод 60) ЕМА 
df['MACD_12_60'] = df['ema_12'] - df['ema_60']
# MACD между средне-быстрой (период 24) и медленной (перирод 120) ЕМА
df['MACD_24_120'] = df['ema_24'] - df['ema_120']
# MACD между самой быстрой (период 12) и медленной (перирод 120) ЕМА
df['MACD_12_120'] = df['ema_12'] - df['ema_120']
# MACD между самой быстрой (период 12) и средне-быстрой (перирод 24) ЕМА
df['MACD_12_24'] = df['ema_12'] - df['ema_24']
df
```

### Создание RSI

```python
# добавляем признак "close_prev": предыдущее значение close
df['close_prev'] = df['close'].shift(1)
df.head(5)
```

```python
df = df[1:]
df = df.reset_index(drop=True)
df
```

```python
# признак типичная цена: (H + L + С) /3
df['tipical_price'] = (df['high'] + df['low'] + df['close'])/3
df['tipical_price'][:5]
```

```python
# признак длина тела свечи
df['body_len'] = df['close'] - df['open']
df.tail(5)
```

```python
# на сколько % цена выросла на бычьей свече
def diff_percent_up(row):
    if row['body_len'] > 0:
        return row['body_len'] / row['close_prev']
    else:
        return 0
```

```python
# признак с только положительными разницами с пред закрытием для RSI (положит значение или 0)
df['close_up'] = df.apply(diff_percent_up, axis=1)
df.tail(5)
```

```python
# описание признака close_up
df['close_up'].describe()
```

```python
# на сколько % цена снизилась на медвежьей свече
def diff_percent_down(row):
    if row['body_len'] < 0:
        return abs(row['body_len']) / row['close_prev']
    else:
        return 0
```

```python
# признак с только отрицательными разницами с пред закрытием для RSI (отрицательное значение или 0)
# значения берем по модулю, все положительные или 0
#df['close_down'] = df['body_len'].apply(lambda x: abs(x) if x<0 else 0)
df['close_down'] = df.apply(diff_percent_down, axis=1)
df.tail(5)
```

```python
# описание признака close_down
df['close_down'].describe()
```

Экспоненциально сглаженное скользящее среднее определяется путем добавления к предыдущему значению скользящего среднего определенной доли текущей цены закрытия. При использовании экспоненциальных скользящих средних больший вес имеют последние цены закрытия. Р-процентное экспоненциальное скользящее среднее:

EMA = (CLOSE (i) * P) + (EMA (i - 1) * (1 - P)) 

где:

CLOSE (i) — цена закрытия текущего периода;

EMA (i - 1) — значение скользящего среднего предыдущего периода;

P — доля использования значения цен.

```python
# n - количество периодов
n = 28
# smooth - сглаживание (числитель), чем он больше, тем больше веса текущему значению
smooth = 2
# сглаживающий фактор
a = smooth/(1+n)
```

```python
# создает список значений ema с параметрами N периодов, сглаживанием smooth и по значениям стобца col_name
def create_ema(n, smooth, col_name):
    a = smooth/(1+n)
    ema = list()
    for i in list(range(0, df.shape[0])):
        if i == 0:
            ema.append(df[col_name][i])
        else:
            ema_val = (df[col_name][i] * a) + (ema[i-1] * (1-a))
            ema.append(ema_val)
    return ema
```

```python
# EMA up
df['ema_up'] = pd.Series(create_ema(n, 1, 'close_up'))
df.head(5)
```

```python
# EMA down
df['ema_down'] = pd.Series(create_ema(28, 1, 'close_down'))
df.head(5)
```

```python
# RS - отношение растущих свечей к убывающим
# новый признак RS  - отношение растущих свечей к убывающим (всегда положительный)
df['rs'] = df['ema_up'] / df['ema_down']
df[['ema_up', 'ema_down', 'rs']]
```

```python
df['rs'].describe()
```

Индикатор вычисляется с помощью экспоненциального сглаживания: находятся средние значения минимумов и максимумов цены.

Индекс относительной силы считаем по формуле:

RSI = 100 – 100 / (1 + RS),

где RS — отношение среднего роста цены, взятой за период, к среднему снижению цены за тот же период, то есть

RS = EMAn(Up) / EMAn(Down)

Информация предоставлена: https://alfaforex.ru/faq/instrumenty-ta/indeks-otnositelnoy-sily/
 

```python
# создание rsi
df['rsi'] = df['rs'].apply(lambda x: 100-100/(1+x))
df[['rs', 'rsi']][:10]
```

### Momentum

```python
# сдвиги для моментума
momentum_shift = 12
# моментум цены по закрытию час назад (сдвиг 12 баров)
df['mom_shift_price'] = df['close'].shift(momentum_shift)
# моментум медленной ЕМА в последний час
df['mom_shift_ema_120_12'] = df['ema_120'].shift(momentum_shift)
# тренд по медленной ЕМА в последние 5 часов
df['mom_shift_ema_120_60'] = df['ema_120'].shift(60)
# тренд по медленной ЕМА в последние 10 часов
df['mom_shift_ema_120_120'] = df['ema_120'].shift(120)
# моментум средней ЕМА в последний час
df['mom_shift_ema_60_12'] = df['ema_60'].shift(momentum_shift)
# моментум средне-быстрой ЕМА в последний час
df['mom_shift_ema_24_12'] = df['ema_24'].shift(momentum_shift)
# моментум быстрой ЕМА в последний час
df['mom_shift_ema_12_12'] = df['ema_12'].shift(momentum_shift)
# моментум средней ЕМА в последние 3 часа
df['mom_shift_ema_60_36'] = df['ema_60'].shift(momentum_shift*3)
# тренд по cупер-медленной ЕМА в последние 5 часов
df['mom_shift_ema_420_60'] = df['ema_420'].shift(60)
df = df[420:]
```

```python
# моментум цены по закрытию час назад
def create_momentum(row):
    return (row['close']- row['mom_shift_price'])/row['close']    
df['momentum_price'] = df.apply(create_momentum, axis=1)

# моментум медленной ЕМА в последний час
def create_mom_ema_120_12(row):
    return (row['ema_120'] - row['mom_shift_ema_120_12'])/row['ema_120']
df['momentum_ema_120_12'] = df.apply(create_mom_ema_120_12, axis=1)

# тренд по медленной ЕМА в последние 5 часов
def create_mom_ema_120_60(row):
    return (row['ema_120'] - row['mom_shift_ema_120_60'])/row['ema_120']
df['momentum_ema_120_60'] = df.apply(create_mom_ema_120_60, axis=1)

# тренд по медленной ЕМА в последние 10 часов
def create_mom_ema_120_120(row):
    return (row['ema_120'] - row['mom_shift_ema_120_120'])/row['ema_120']
df['momentum_ema_120_120'] = df.apply(create_mom_ema_120_120, axis=1)

# моментум средней ЕМА в последний час
def create_mom_ema_60_12(row):
    return (row['ema_60'] - row['mom_shift_ema_60_12'])/row['ema_60']
df['momentum_ema_60_12'] = df.apply(create_mom_ema_60_12, axis=1)

# моментум средне-быстрой ЕМА в последний час
def create_mom_ema_24_12(row):
    return (row['ema_24'] - row['mom_shift_ema_24_12'])/row['ema_24']
df['momentum_ema_24_12'] = df.apply(create_mom_ema_24_12, axis=1)

# моментум быстрой ЕМА в последний час
def create_mom_ema_12_12(row):
    return (row['ema_12'] - row['mom_shift_ema_12_12'])/row['ema_12']
df['momentum_ema_12_12'] = df.apply(create_mom_ema_12_12, axis=1)

# моментум средней ЕМА в последние 3 часа
def create_mom_ema_60_36(row):
    return (row['ema_60'] - row['mom_shift_ema_60_36'])/row['ema_60']
df['momentum_ema_60_36'] = df.apply(create_mom_ema_60_36, axis=1)

# моментум супер-медленной ЕМА в последние 5 часов
def create_mom_ema_420_60(row):
    return (row['ema_420'] - row['mom_shift_ema_420_60'])/row['ema_420']
df['momentum_ema_420_60'] = df.apply(create_mom_ema_420_60, axis=1)

df.head()[df.columns[-15:]]
```

## Cтратегия: вход на 4й бар после 50%

Вход на закрытии третьей свечи после ухода индекса выше 50%:

    - если индекс поднялся выше уровня 50%,то
    
        - если индекс остается выше 50%, 
        
            - то заходим позакрытию четвертой такой свечи

```python
# индекс первой свечи, где RSI уходит выше 50%
ind_first = rsi_low_index[0]
# индекс последней свечи, где RSI выше 50%, может совпадать с ind_first
ind_last = 0
# (ind_last - ind_first) +1 - длительность нахождения инднкса выше 50% в барах
# список первых индексов
ind_first_list = list()
# добавляем в него первый индекс
ind_first_list.append(ind_first)
# список последних индексов (должны совпадать по длине, значения могут быть в обоих списках)
ind_last_list = list()

for i in list(range(1, len(rsi_low_index))):
    # если текущий индекс больше предыдущего более, чем на 1 (идут не подряд)
    if rsi_low_index[i] > (rsi_low_index[i-1]+1):
        # предыдущий индекс - это последний в серии подряд идущих
        ind_last = rsi_low_index[i-1]
        # добавляем его в список последних индексов
        ind_last_list.append(ind_last)
        # текущий инд - это первый в серии подряд идущих индексов
        ind_first = rsi_low_index[i]
        # добавляем его в список первых индексов
        ind_first_list.append(ind_first)
    # если текущий и предыдущий индексы идут подряд (отличаются на 1), 
    # то ничего не делаем
if len(ind_first_list) > len(ind_last_list):
    ind_first_list.pop()
    
print('Последние индексы длина: ', len(ind_last_list), '\n')    
print('Первые индексы: ', ind_first_list[:10], '\n')
print('Последние индексы: ', ind_last_list[:10])
```

#### Добавляем апсайд/даунсайд по всем сделкам

Для всех паттернов 4 свечи считаем апсайд/даунсайд, считаем корреляцию апсайда/даунсайда с вводными данными: берем выборку из датафрейма, где паттерн 4 и более свечи и смотрим корреляцию с получившимся списком ап/даунсайдов.

Смотрим общую картину без фильтров: хуже нее выборки быть не должны.

```python
delay = 4
in_indexes_filt_delay = in_out_indexes[in_out_indexes['duration'] >= delay].index+(delay-1)

in_indexes_filt_delay
```

```python
# считаем апсайд и даунсайд 
# сигнал: пересечение RSI уровня 50% снизу вверх
upside_list = list()
downside_list = list()
max_rsi_list = list()

# для каждого индекса выхода за 50% снизу вверх
for i in (in_indexes_filt_delay):
    # цена открытия сделки по закрытию первой свечи после пересечения
    buy = train.loc[i]['close']
    # берем длительность нахождения RSI выше уровня в барах и по ней считаем закрытие (окончание паттерна)
    # i-2 т к вход по закрытию третьего бара, т е открытие было на 2 меньше
    # из длительности вычитаем 3, т к к заходу три бара уже пропущено
    close_bar_ind = in_out_indexes[in_out_indexes['in'] == i-(delay-1)]['out'].values[0]
    # upside - максимум в первый час
    high = train.loc[i+1:close_bar_ind+1]['high'].max() - buy
    upside_list.append(high)
    low = train.loc[i+1:close_bar_ind+1]['low'].min() - buy
    downside_list.append(low)
    rsi = train.loc[i+1:close_bar_ind+1]['rsi'].max()
    max_rsi_list.append(rsi)
    
display('upside:', pd.Series(upside_list).describe())
display('downside:', pd.Series(downside_list).describe())
display('max_rsi:', pd.Series(max_rsi_list).describe())
display((pd.Series(upside_list).describe()/pd.Series(downside_list).describe())[['50%', 'mean']])
```

В данном виде стратегия абсолютно не жизнеспособна даже без учета издержек на комиссии и спреды: медианные апсайды и даунсайды равны.

```python
# считаем апсайд и даунсайд 
# сигнал: пересечение RSI уровня 50% снизу вверх
upside_list = list()
downside_list = list()
tipical_price_list = list()
duration_list = list()
# максимальный rsi за время нахождения индекса выше 50%
max_rsi_list = list()

# для каждого индекса выхода за 20% снизу вверх
for i in (rsi_low_index_str_2):
    # цена открытия сделки по закрытию первой свечи после пересечения
    buy = train.loc[i]['close']
    # берем длительность нахождения RSI выше уровня в барах и по ней считаем закрытие (окончание паттерна)
    # i-2 т к вход по закрытию третьего бара, т е открытие было на 2 меньше
    # из длительности вычитаем 3, т к к заходу три бара уже пропущено
    close_bar_ind = in_out_indexes[in_out_indexes['in'] == i-(delay-1)]['out'].values[0]
    # upside - максимум в первый час
    high = train.loc[i+1:close_bar_ind+1]['high'].max() - buy
    upside_list.append(high)
    low = train.loc[i+1:close_bar_ind+1]['low'].min() - buy
    downside_list.append(low)
    tipical_price = train.loc[close_bar_ind]['close'] - buy
    tipical_price_list.append(tipical_price)
    rsi = train.loc[i+1:close_bar_ind+1]['rsi'].max()
    max_rsi_list.append(rsi)
    duration_list.append(in_out_indexes[in_out_indexes['in'] == i-(delay-1)]['duration'].values[0])
```
